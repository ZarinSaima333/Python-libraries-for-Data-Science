````python
import pandas as pd

# Create a sample DataFrame
data = {'column_name': ['1', '2', '3', '4', '5', 'CarlSarah']}
df = pd.DataFrame(data)

# Convert the column to numeric, coercing errors to NaN
df['column_name'] = pd.to_numeric(df['column_name'], errors='coerce')

# Calculate the mean
mean_value = df['column_name'].mean()

print("Mean:", mean_value)
````

````python
df
````

````python
import pandas as pd

# Create the DataFrame
data = {
    "company": ["GOOG", "GOOG", "MSFT", "MSFT", "FB", "FB"],
    "person": ["Sam", "Charlie", "Amy", "Vannesa", "Carl", "Sarah"],
    "Sales": [200, 120, 340, 124, 243, 350]
}

df = pd.DataFrame(data)

# Calculate the mean sales for each company
mean_sales = df.groupby('company')['Sales'].mean()

print(mean_sales)
````

````python
sum_sales = df.groupby('company')['Sales'].sum()
sum_sales
````

````python
std_sales = df.groupby('company')['Sales'].std()
std_sales
````

##Difficult Example

````python
left=pd.DataFrame({"key1":["K0","K0",'K1','K2'],
                   "Key2":['K0','K1','K0','K1'],
                   "A":["A0",'A1','A2','A3'],
                  "B":["B0",'B1','B2','B3']},index=[11,12,13,14])

right=pd.DataFrame({"key1":["K0","K1",'K1','K2'],
                  "Key2":['K0','K0','K0','K0'],
                  "C":["C0",'C1','C2','C3'],
                   "D":["D0",'D1','D2','D3']},index=[11,12,13,14])
````

````python
left
````

````python
right
````

````python
pd.merge(left,right,on=["key1",'Key2'])
````

Matching rows: The merge operation looks for rows where the values in the "key1" and "Key2" columns are the same in both DataFrames.

````python
pd.merge(left,right,how="outer",on=["key1",'Key2'])
````

````python
pd.merge(left,right,how="right",on=["key1",'Key2'])
````

````python
pd.merge(left,right,how="left",on=["key1",'Key2'])
````

##Joining

joining is a convenient method for combining the columns of two potentially different-indexed DataFrames into a single result DataFrame.


keys you want to join on are actually index instead of column.

````python
left=pd.DataFrame({"A":["A0",'A1','A2'],
                   "B":["B0",'B1','B2']},
                  index=['K0','K1','K2'])
right=pd.DataFrame({"C":["C0",'C2','C3'],
                   "D":["D0",'D2','D3']},
                  index=['K0','K2','K3'])
````

````python
left.join(right) #left full+right joto tuku match
````

````python
right.join(left)
````

````python
left.join(right,how="inner") #Nan chara
````

````python
left.join(right,how="outer") #Naan shoho
````
